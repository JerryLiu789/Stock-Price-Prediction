**股價爬蟲**

```python
import requests
import pandas as pd
from io import StringIO

# 台灣證交所每日收盤行情 https://www.twse.com.tw/zh/trading/historical/mi-index.html

# 爬蟲每日收盤行情
response = requests.get('https://www.twse.com.tw/rwd/zh/afterTrading/MI_INDEX?date=20240701&type=ALLBUT0999&response=csv&_=1719884229645')

# 查看資料狀況
lines = response.text.split('\n')
lines[:10]
```

```python
# 資料格式整理-確認要使用的資料格式長度為何
for line in lines:
  print(len(line.split('",')))
```

```python
# 篩選長度為17的資料加入newlines
newlines = []
for line in lines:
  if len(line.split('",')) == 17:
    newlines.append(line)
newlines[:10]
```

```python
# 將資料轉換為DataFrame並且處理格式
df = pd.read_csv(StringIO('\n'.join(newlines).replace('=','')))
df = df.astype(str)
df = df.apply(lambda x: x.str.replace(',',''))
df = df.set_index('證券代號')
df = df.apply(lambda x: pd.to_numeric(x, errors = 'coerce'))
df = df[df.columns[df.isnull().sum() != len(df)]]

df.head()
```

```python
# 功能測試，確認當日漲幅超過5%的股票
df[df['收盤價']/df['開盤價']>1.05]
```

```python
# 功能測試，存成CSV檔
df.to_csv('daily_price.csv', encoding = 'utf_8_sig')

# 功能測試，讀取CSV檔
df = pd.read_csv('daily_price.csv',index_col = ['證券代號'])
df.head()
```

```python
# 功能測試，存到 sqlite3 中
import sqlite3
conn = sqlite3.connect('test.sqlite3')
df.to_sql('daily_price', conn, if_exists = 'replace')
```

```python
# 功能測試，讀取 sqlite3 內容
df = pd.read_sql('select * from daily_price', conn, index_col=['證券代號'])
df.head()
```
