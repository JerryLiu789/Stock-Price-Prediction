**財報爬蟲**

```python
import requests
from io import StringIO
import pandas as pd

# IFRS 合併財務報表 https://mops.twse.com.tw/server-java/t164sb01?step=1&CO_ID=1101&SYEAR=2017&SSEASON=3&REPORT_ID=C
# CO_ID: 1101 （股票代號）
# SYEAR: 2017 （年度）
# SSEASON: 3 （季度）
# REPORT_ID: C （合併財務報表）


res = requests.get('https://mops.twse.com.tw/server-java/t164sb01?step=1&CO_ID=1101&SYEAR=2017&SSEASON=3&REPORT_ID=C')
res.encoding = 'big5'

# 將html轉換為檔案
dfs = pd.read_html(StringIO(res.text))
dfs[1]
```

```python
import time
import os

# 檔案儲存位置，如果 financial_report 資料夾不存在，則創健這個資料夾
if 'financial_report' not in os.listdir():
  os.mkdir('financial_report')

# 選擇要爬的股票代號
sid = ['1101','2330']

# 對於要爬的股票代號進行爬蟲並儲存
for s in sid:
  res = requests.get('https://mops.twse.com.tw/server-java/t164sb01?step=1&CO_ID=' + s + '&SYEAR=2017&SSEASON=3&REPORT_ID=C')
  res.encoding = 'big5'

  # 設定存檔的路徑
  path = os.path.join('financial_report', s + '.html')

  # 將檔案打開，寫入html，然後關閉
  f = open(path,'w',encoding='utf-8')
  f.write(res.text)
  f.close()

  print(s)

  # 設定休息時間，避免過度與 Sever 交互給 Sever 過多負擔被封鎖
  time.sleep(10)
```

```python
dfs = []

# 讀取檔案
for s in sid:
  path = os.path.join('financial_report', s + '.html')

  # 將檔案存於 dfs 中
  dfs.append(pd.read_html(path,encoding='utf-8'))

# 匯出 2330 的第一個 DataFrame
dfs[1][1].head(10)
```
